```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('/content/mtcars (2).csv')
print(df)
sns.scatterplot(x='wt', y='mpg', data=df)
plt.xticks(rotation=90)
plt.xlabel('mpg')
plt.ylabel('wt')
plt.title('scatterplot: frequency of mpg')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/mtcars (2).csv')
print(df)
plt.figure(figsize=(8, 5))
plt.hist(df['mpg'], bins=9, edgecolor='black', alpha=0.7)
plt.xlabel('Miles per gallon (mpg)')
plt.ylabel('Frequency')
plt.title('Histogram of Miles per gallon (mpg)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/mtcars (2).csv')
print(df)
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['mpg'])
plt.xlabel('Miles per gallon (mpg)')
plt.title('Box Plot of Miles per gallon (mpg)')
plt.grid(True)
plt.show()
```
